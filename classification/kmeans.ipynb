```python
# k-means algorithm

from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# generate random data

X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
# train model

clf = KMeans(n_clusters=3, random_state=0)
clf.fit(X)
```

```python
plt.figure(figsize=(10, 10))
plt.title('K-Means')
plt.xlabel('x1')
plt.ylabel('x2')

# plot the data points with a different color for each cluster
predictions = clf.predict(X)
plt.scatter(X[:,0], X[:,1], c=predictions, s=50, cmap='viridis')

# plot the centroids
centroids = clf.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='x', c='red', s=200, linewidths=3)

plt.show()
```
