```python
# cubic regression using gradient descent without matrices

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('data/cubic.xlsx')

x_train = data['x'].values
y_train = data['y'].values
```

```python
theta0 = 0
theta1 = 0
theta2 = 0
theta3 = 0

alpha = 0.1
m = len(x_train)
```

```python
for _ in range(100000):
    # tweak the intercept
    losses = 0
    for i in range(len(x_train)):
        h = theta0 + theta1 * x_train[i] + theta2 * x_train[i] ** 2 + theta3 * x_train[i] ** 3
        loss = h - y_train[i]
        losses += loss
    temp0 = theta0 - alpha / m * losses

    # tweak theta1
    losses = 0
    for i in range(len(x_train)):
        h = theta0 + theta1 * x_train[i] + theta2 * x_train[i] ** 2 + theta3 * x_train[i] ** 3
        loss = (h - y_train[i]) * x_train[i]
        losses += loss
    temp1 = theta1 - alpha / m * losses

    # tweak theta2
    losses = 0
    for i in range(len(x_train)):
        h = theta0 + theta1 * x_train[i] + theta2 * x_train[i] ** 2 + theta3 * x_train[i] ** 3
        loss = (h - y_train[i]) * x_train[i] ** 2
        losses += loss
    temp2 = theta2 - alpha / m * losses

    # tweak theta3
    losses = 0
    for i in range(len(x_train)):
        h = theta0 + theta1 * x_train[i] + theta2 * x_train[i] ** 2 + theta3 * x_train[i] ** 3
        loss = (h - y_train[i]) * x_train[i] ** 3
        losses += loss
    temp3 = theta3 - alpha / m * losses

    # update values
    theta0 = temp0
    theta1 = temp1
    theta2 = temp2
    theta3 = temp3
```

```python
# print equation
print('y = %.2f + %.2fx + %.2fx^2 + %.2fx^3' % (theta0, theta1, theta2, theta3))
```

```python
# plot results
plt.figure(figsize=(10, 10))
plt.title('Cubic Regression')
plt.xlabel('x')
plt.ylabel('y')

y_pred = theta0 + theta1 * x_train + theta2 * x_train ** 2 + theta3 * x_train ** 3
plt.plot(x_train, y_pred, 'k-')
plt.plot(x_train, y_train, 'o')

plt.show()
```
